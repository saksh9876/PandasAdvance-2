Q2. What is the difference between the functions loc and iloc in pandas.DataFrame?

ans:
    
    
    
    The `loc` and `iloc` are both accessors in pandas.DataFrame, but they are used for different purposes:

1. `loc`:
   - `loc` is primarily label-based indexing. It is used to select rows and columns from a DataFrame based on their labels (index and column names).
   - When using `loc`, you specify the row label(s) and column label(s) you want to access, separated by a comma.
   - The labels can be either single labels, a list of labels, or a slice object.
   - For example, `df.loc[2]` will give you the row with the label/index 2, and `df.loc[:, 'column_name']` will give you all rows of the specific column.

2. `iloc`:
   - `iloc` is primarily integer-based indexing. It is used to select rows and columns from a DataFrame based on their integer positions (row and column numbers).
   - When using `iloc`, you specify the row position(s) and column position(s) you want to access, separated by a comma.
   - The positions are zero-based, meaning the first row or column has a position of 0, the second row or column has a position of 1, and so on.
   - For example, `df.iloc[2]` will give you the third row of the DataFrame (row index 2), and `df.iloc[:, 1]` will give you the second column of the DataFrame (column index 1).

Here's a comparison of the usage of `loc` and `iloc`:

```python
import pandas as pd

data = {
    'A': [10, 20, 30],
    'B': [40, 50, 60],
    'C': [70, 80, 90]
}

df = pd.DataFrame(data, index=['X', 'Y', 'Z'])

# Using loc (label-based indexing)
print(df.loc['Y'])        # Access row with label 'Y'
print(df.loc[['X', 'Z']]) # Access rows with labels 'X' and 'Z'
print(df.loc[:, 'B'])     # Access column 'B'
print(df.loc['X', 'C'])   # Access element at row 'X' and column 'C'

# Using iloc (integer-based indexing)
print(df.iloc[1])          # Access row at position 1 (second row)
print(df.iloc[[0, 2]])     # Access rows at positions 0 and 2
print(df.iloc[:, 1])       # Access column at position 1 (second column)
print(df.iloc[0, 2])       # Access element at row position 0 and column position 2
```

Both `loc` and `iloc` are powerful tools for accessing specific data within a DataFrame. The choice between `loc` and `iloc` depends on whether you want to access data based on labels or integer positions.

Q6. What do you understand about the windows function in pandas and list the types of windows
functions?

ans:
    
    
    In pandas, the term "window functions" refers to a class of functions that operate on a specific window (subset) of data within a DataFrame. These functions allow you to perform calculations and manipulations on a defined window of data points instead of operating on the entire DataFrame at once. Window functions are powerful tools for performing operations that involve looking at a group of rows in relation to a current row.

The two main types of window functions in pandas are:

1. Rolling Windows:
   - Rolling windows are used to create a rolling view over a specified number of rows in a DataFrame.
   - The rolling window moves along the DataFrame, and for each position, it considers a fixed-size window of data.
   - Common window operations include calculating moving averages, calculating rolling sums, and applying custom functions to a sliding window of data.
   - You can use the `rolling()` function to create a rolling window object, and then you can apply various aggregation functions like `mean()`, `sum()`, etc.

2. Expanding Windows:
   - Expanding windows gradually include more rows over time as the window grows.
   - Expanding windows are used to compute cumulative values over time.
   - For each row, an expanding window function considers all the rows from the beginning of the DataFrame up to that row.
   - You can use the `expanding()` function to create an expanding window object, and then you can apply various aggregation functions like `mean()`, `sum()`, etc.

Here's a brief example of using a rolling window to calculate the rolling mean of a DataFrame:

```python
import pandas as pd

data = {
    'Value': [10, 20, 30, 40, 50]
}

df = pd.DataFrame(data)

# Calculate the rolling mean using a window of size 3
rolling_mean = df['Value'].rolling(window=3).mean()
print(rolling_mean)
```

Output:

```
0     NaN
1     NaN
2    20.0
3    30.0
4    40.0
Name: Value, dtype: float64
```

In this example, we use the `rolling()` function on the 'Value' column with `window=3` to specify a rolling window of size 3. The result is the rolling mean of the 'Value' column, considering each window of three rows. The first two rows have NaN (Not a Number) as there are not enough rows to calculate the mean, but from the third row onwards, the rolling mean is computed.

Q8. Write a Python program that takes in two dates as input (in the format YYYY-MM-DD) and
calculates the difference between them in days, hours, and minutes using Pandas time delta. The

ans:
    
    You can use pandas' `pd.to_datetime()` function to convert the input dates to pandas datetime objects and then use the pandas timedelta to calculate the difference between them. Here's the Python program to achieve this:

```python
import pandas as pd

def get_date_input():
    while True:
        try:
            date_str = input("Enter a date in the format YYYY-MM-DD: ")
            date_obj = pd.to_datetime(date_str)
            return date_obj
        except ValueError:
            print("Invalid date format. Please enter the date in the format YYYY-MM-DD.")

# Get input dates from the user
print("Enter the first date:")
date1 = get_date_input()

print("\nEnter the second date:")
date2 = get_date_input()

# Calculate the difference between the dates
time_difference = date2 - date1

# Display the result
print("\nDifference between the dates:")
print(f"Days: {time_difference.days}")
print(f"Hours: {time_difference.seconds // 3600}")
print(f"Minutes: {(time_difference.seconds % 3600) // 60}")
```

The timedelta between the two input dates is calculated using the `-` operator. The result is then displayed, showing the difference in days, hours, and minutes.



Q9. Write a Python program that reads a CSV file containing categorical data and converts a specified
column to a categorical data type. The program should prompt the user to enter the file path, column
name, and category order, and then display the sorted data.

ans:
    
    To achieve this, we can use pandas to read the CSV file, convert the specified column to a categorical data type, and then sort the data based on that column. Here's the Python program:

```python
import pandas as pd

def read_csv_and_convert_to_categorical(file_path, column_name, category_order):
    try:
        # Read the CSV file into a DataFrame
        df = pd.read_csv(file_path)

        # Convert the specified column to a categorical data type
        df[column_name] = pd.Categorical(df[column_name], categories=category_order, ordered=True)

        # Sort the DataFrame based on the categorical column
        sorted_df = df.sort_values(by=column_name)

        return sorted_df

    except FileNotFoundError:
        print(f"File not found at path: {file_path}")
        return None
    except KeyError:
        print(f"Column not found: {column_name}")
        return None

# Prompt the user to enter file path, column name, and category order
file_path = input("Enter the CSV file path: ")
column_name = input("Enter the column name to convert to categorical: ")
category_order = input("Enter the category order separated by spaces: ").split()

# Convert the input category order to a list of strings
category_order = [str(item) for item in category_order]

# Read the CSV file, convert the column to categorical, and sort the data
sorted_data = read_csv_and_convert_to_categorical(file_path, column_name, category_order)

if sorted_data is not None:
    print("\nSorted Data:")
    print(sorted_data)
```

Q10. Write a Python program that reads a CSV file containing sales data for different products and
visualizes the data using a stacked bar chart to show the sales of each product category over time. The
program should prompt the user to enter the file path and display the chart.

ans:
    
    
    To visualize the sales data from a CSV file using a stacked bar chart, we can use pandas to read the CSV file and matplotlib to create the plot. Here's the Python program:

```python
import pandas as pd
import matplotlib.pyplot as plt

def visualize_sales_data(file_path):
    try:
        # Read the CSV file into a DataFrame
        df = pd.read_csv(file_path)

        # Convert the 'Date' column to datetime format
        df['Date'] = pd.to_datetime(df['Date'])

        # Group the data by 'Date' and 'Product Category' and calculate total sales
        sales_data = df.groupby(['Date', 'Product Category'])['Sales'].sum().unstack()

        # Create a stacked bar chart
        ax = sales_data.plot(kind='bar', stacked=True, figsize=(10, 6))
        plt.title('Sales of Each Product Category Over Time')
        plt.xlabel('Date')
        plt.ylabel('Sales')
        plt.legend(title='Product Category', loc='upper left')

        # Display the chart
        plt.show()

    except FileNotFoundError:
        print(f"File not found at path: {file_path}")
    except KeyError:
        print("CSV file does not contain required columns: 'Date', 'Product Category', and 'Sales'.")

# Prompt the user to enter the file path
file_path = input("Enter the CSV file path: ")

# Visualize the sales data using a stacked bar chart
visualize_sales_data(file_path)
```

```python
import pandas as pd
from tabulate import tabulate

def calculate_statistics(file_path):
    try:
        # Read the CSV file into a DataFrame
        df = pd.read_csv(file_path)

        # Calculate the mean, median, and mode of the test scores
        mean_score = df['Test Score'].mean()
        median_score = df['Test Score'].median()
        mode_scores = df['Test Score'].mode()

        # Convert the mode to a comma-separated string
        mode_str = ', '.join(str(score) for score in mode_scores)

        # Create a table to display the statistics
        table = [['Mean', mean_score], ['Median', median_score], ['Mode', mode_str]]

        # Display the table
        print(tabulate(table, headers=['Statistic', 'Value'], tablefmt='grid'))

    except FileNotFoundError:
        print(f"File not found at path: {file_path}")
    except KeyError:
        print("CSV file does not contain the required column 'Test Score'.")

# Prompt the user to enter the file path
file_path = input("Enter the file path of the CSV file containing the student data: ")

# Calculate and display the statistics
calculate_statistics(file_path)
```
